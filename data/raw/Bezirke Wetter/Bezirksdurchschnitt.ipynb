```python
berliner_bezirke = [
    "Mitte",
    "Friedrichshain-Kreuzberg",
    "Pankow",
    "Charlottenburg-Wilmersdorf",
    "Spandau",
    "Steglitz-Zehlendorf",
    "Tempelhof-Schöneberg",
    "Neukölln",
    "Treptow-Köpenick",
    "Marzahn-Hellersdorf",
    "Lichtenberg",
    "Reinickendorf"
]
```

```python
import pandas as pd
```

```python
dataframes = []
for name in berliner_bezirke:
    df = pd.read_csv('open-meteo-' + name + '.csv', header = 2)
    df = df.set_index('time')
    dataframes.append(df)
```

```python
dataframes[0]
```

```python
average_df = pd.concat(dataframes).groupby(level=0).mean()
```

```python
average_df
```

```python
average_df.to_csv('Bezirke_Durchschnitt.csv')
```
